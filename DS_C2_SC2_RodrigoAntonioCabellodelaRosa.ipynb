#### a. Extracci칩n de informaci칩n de archivo:

```python
import pandas as pd
import numpy as np

vacunacion_mundial = pd.read_csv('/content/country_vaccinations.csv', parse_dates= ['date'])
vacunacion_mundial
```

#### b. Mostrar la estructura y tipos de datos de cada columna para identificar qu칠 operaciones puedes realizar con cada una de ellas, asegur치ndote que las columnas con fechas sean del tipo datetime64.

#### c. Determinar la cantidad de vacunas aplicadas de cada compa침칤a (con base en c칩mo lo reporta cada pa칤s en la columna vaccines, en otras palabras, agrupe por vaccines y realice la sumatoria):

```python
vacunas_suma = vacunacion_mundial[['daily_vaccinations', 'vaccines']].groupby(['vaccines']).sum()
vacunas_suma
```

---
#### d. Obtener la cantidad de vacunas aplicadas en todo el mundo.

```python
vacunas_total = vacunacion_mundial['daily_vaccinations'].sum()
print('Total de vacunas aplicadas en todo el mundo:', vacunas_total)
```

---
#### e. Calcular el promedio de vacunas aplicadas por pa칤s.

```python
vacunas_pais = vacunacion_mundial[['daily_vaccinations', 'country']].groupby('country').mean()
vacunas_pais
```

---
### f. Determinar la cantidad de vacunas aplicadas el d칤a *29/01/21* en todo el mundo.

```python
vacunas_dia = vacunacion_mundial.loc[vacunacion_mundial['date']== '2021-01-21', ['daily_vaccinations','date']]
print('Total de vacunas el d칤a 21-01-2024:', vacunas_dia['daily_vaccinations'].sum())
```

---
#### g. Crear un dataframe nuevo denominado conDiferencias que contenga los datos originales y una columna derivada (diferencias) con las diferencias de aplicaci칩n entre las columnas daily_vaccionations y daily_vaccionations_raw.  

```python
vacunacion_mundial['diferencias'] = vacunacion_mundial['daily_vaccinations'] - vacunacion_mundial['daily_vaccinations_raw']
conDiferencias = vacunacion_mundial.copy()
conDiferencias
```

---
#### h. Obtener el periodo de tiempo entre el registro con fecha m치s reciente y el registro con fecha m치s antigua.

```python
registroTiempo = vacunacion_mundial['date'].max() - vacunacion_mundial['date'].min()
print('Tiempo entre primer y 칰ltimo registro:', registroTiempo)
```

---
#### i. Crear un dataframe nuevo denominado conCantidad que contenga los datos originales y una columna derivada (canVac) con la cantidad de vacunas utilizadas cada d칤a (usar la columna vaccines y separar por el car치cter , ).

```python
vacunacion_mundial['canVac']= vacunacion_mundial['vaccines'].str.split(",").str.len()
conCantidad = vacunacion_mundial.copy()
conCantidad
```

---
#### j. Generar un dataframe denominado antes20 con todos los registros que se hayan realizado antes del 20 de diciembre de 2020.

```python
antes20 = vacunacion_mundial[vacunacion_mundial['date'] < '2020-12-20']
antes20
```

---
#### k. Obtener un dataframe denominado pfizer con todos los registros donde se haya utilizado la vacuna Pfizer.

```python
pfizer = vacunacion_mundial[vacunacion_mundial['vaccines'].str.contains('Pfizer', case=False, na=False)]
pfizer
```

---
#### l. Almacenar los dataframes generados (conDiferencias, conCantidad, antes20 y pfizer) en un archivo de Excel denominado resultadosReto.xlsx, donde cada dataframe ocupe una hoja diferente.

```python
with pd.ExcelWriter('/content/resultadosReto.xlsx') as writer:
    conDiferencias.to_excel(writer, sheet_name= "conDiferencias")
    conCantidad.to_excel(writer, sheet_name="conCantidad")
    antes20.to_excel(writer, sheet_name="antes20")
    pfizer.to_excel(writer, sheet_name="pfizer")
```
